# **1. Perkenalan Dataset**

## Dataset Credit Scoring

Dataset ini berisi data credit scoring yang digunakan untuk memprediksi risiko kredit (target). Dataset memiliki 300 baris data dengan 8 fitur.

**Sumber Dataset:** Dataset ini dibuat untuk keperluan eksperimen credit scoring.

**Penjelasan Fitur:**
- `age`: Usia nasabah
- `income`: Pendapatan nasabah
- `loan_amount`: Jumlah pinjaman
- `employment_years`: Lama bekerja dalam tahun
- `home_ownership`: Status kepemilikan rumah (rent, own, mortgage)
- `marital_status`: Status pernikahan (single, married)
- `late_payments`: Jumlah pembayaran terlambat
- `target`: Target risiko kredit (0 = rendah risiko, 1 = tinggi risiko)

# **5. Data Preprocessing**

```python
# Preprocessing manual:
# 1. Mengisi missing values pada kolom numerik dengan median
# 2. Mengisi missing values pada kolom kategorikal dengan modus
# 3. Normalisasi format kategorikal

processed = df.copy()

# Preprocessing kolom numerik
for col in ['age','income','loan_amount','employment_years','late_payments']:
    processed[col] = pd.to_numeric(processed[col], errors='coerce')
    processed[col] = processed[col].fillna(processed[col].median())

# Preprocessing kolom kategorikal
for col in ['home_ownership','marital_status']:
    processed[col] = processed[col].astype('string').fillna('unknown').str.strip().str.lower()
    mode_value = processed[col].mode().iloc[0] if not processed[col].mode().empty else 'unknown'
    processed[col] = processed[col].replace({'<NA>': mode_value}).fillna(mode_value)

processed.head()
```

```python
# Simpan data hasil preprocessing
out_path = Path('namadataset_preprocessing/credit_scoring_preprocessing.csv')
out_path.parent.mkdir(parents=True, exist_ok=True)
processed.to_csv(out_path, index=False)
print(f'Data preprocessing disimpan di: {out_path}')
print(f'Jumlah baris: {len(processed)}')
print(f'Jumlah kolom: {len(processed.columns)}')
```
